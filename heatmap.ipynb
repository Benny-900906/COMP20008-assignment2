```python
import seaborn as sns
import pandas as pd
import math

TARGET_FILE = "final.csv"
TEN_K = 10000

file = pd.read_csv(TARGET_FILE, encoding = "ISO-8859-1")

data = file.loc[file["VCE Median Study Score"] >= 0]

data = data.loc[data["Median Income in Postcode"] >= 0]

VCE_score = data["VCE Median Study Score"]

score_40_over = data["Percentage of study scores of 40 and over"]

median_income = list(data["Median Income in Postcode"])
    

df = pd.DataFrame({"Median Income":median_income, "VCE Score": VCE_score, "Score over 40": score_40_over})

df = df.set_index("Median Income")

df = df.sort_values(by = "Median Income", ascending = False)

result = sns.heatmap(df, cmap="jet")
```
